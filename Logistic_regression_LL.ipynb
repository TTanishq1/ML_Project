```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, roc_curve, classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python

# Replace any instances of '#NUM!' with NaN
data.replace('#NUM!', np.nan, inplace=True)

# Convert all columns to numeric, forcing errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values (if you prefer imputation, you can do that here instead)
data.dropna(inplace=True)
```

```python
data['is_safe'] = data['is_safe'].astype(int)
data.dtypes
```

```python
X = data.iloc[:, :-1]  # Features
y = data.iloc[:, -1]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 4. Correlation Matrix with Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# 5. Histograms for Feature Distributions
data.hist(figsize=(15, 12), bins=20)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
LR_model = LogisticRegression(random_state=42, max_iter=500) #+#
LR_model.fit(X_train, y_train)
```

```python
y_pred = LR_model.predict(X_test)
y_pred_prob = LR_model.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print("Logistic Regression Results:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot(cmap='flag')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Get the predicted probabilities for the positive class
y_prob = LR_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Diagonal line for random guess
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

# **DecisionTreeClassifier**

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
y_pred_prob = dt.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print("DecisionTreeClassifier Results:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Plot ROC curve
```

```python
# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot(cmap='flag')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Get the predicted probabilities for the positive class
y_prob = dt.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Diagonal line for random guess
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
# Plotting feature importance
plt.figure(figsize=(10, 8))
feature_importances = pd.Series(dt.feature_importances_, index=X.columns).sort_values(ascending=False)
sns.barplot(x=feature_importances, y=feature_importances.index, palette="viridis")
plt.title("Feature Importance in Decision Tree")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.show()
```

```python
# Plotting the decision tree structure (limited depth for readability)
plt.figure(figsize=(20, 10))
plot_tree(dt, feature_names=X.columns, class_names=["Unsafe", "Safe"], filled=True, max_depth=3, fontsize=10)
plt.title("Decision Tree Structure (Depth = 3 for Visualization)")
plt.show()
```
