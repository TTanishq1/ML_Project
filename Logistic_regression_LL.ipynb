```python

# Replace any instances of '#NUM!' with NaN
data.replace('#NUM!', np.nan, inplace=True)

# Convert all columns to numeric, forcing errors to NaN
data = data.apply(pd.to_numeric, errors='coerce')

# Drop rows with NaN values (if you prefer imputation, you can do that here instead)
data.dropna(inplace=True)
```

```python
data['is_safe'] = data['is_safe'].astype(int)
data.dtypes
```

```python
X = data.iloc[:, :-1]  # Features
y = data.iloc[:, -1]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 4. Correlation Matrix with Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# 5. Histograms for Feature Distributions
data.hist(figsize=(15, 12), bins=20)
plt.suptitle("Feature Distributions")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
LR_model = LogisticRegression(random_state=42, max_iter=500) #+#
LR_model.fit(X_train, y_train)
```

```python
y_pred = LR_model.predict(X_test)
y_pred_prob = LR_model.predict_proba(X_test)[:, 1]
```

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print("Logistic Regression Results:")
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
print("\nClassification Report:\n", classification_report(y_test, y_pred))

```

```python
# Generate and display the confusion matrix
cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot(cmap='flag')
plt.title('Confusion Matrix')
plt.show()
```

```python
# Get the predicted probabilities for the positive class
y_prob = LR_model.predict_proba(X_test)[:, 1]

# Calculate ROC curve and AUC score
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
auc_score = roc_auc_score(y_test, y_prob)

# Plot the ROC curve
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Diagonal line for random guess
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

*Regularizaion*

```python
# Initialize logistic regression model with L1 regularization
model_l1 = LogisticRegression(penalty='l1', C=1.0, solver='liblinear')  # Adjust C for stronger/weaker regularization

# Train the model with L1 regularization
model_l1.fit(X_train, y_train)

# Predict on the test set
y_pred_l1 = model_l1.predict(X_test)
y_prob_l1 = model_l1.predict_proba(X_test)[:, 1]
```

```python
# Display coefficients of features
coef_l1 = model_l1.coef_[0]
feature_importance = pd.DataFrame({'Feature': X.columns, 'Coefficient': coef_l1})

# Show non-zero coefficients
non_zero_features = feature_importance[feature_importance['Coefficient'] != 0]
print("Non-zero features after L1 regularization:\n", non_zero_features)
```

```python
# Confusion Matrix
cm_l1 = confusion_matrix(y_test, y_pred_l1)
ConfusionMatrixDisplay(cm_l1).plot(cmap='flag')
plt.title('Confusion Matrix with L1 Regularization')
plt.show()


```

```python
# ROC Curve and AUC Score
fpr_l1, tpr_l1, _ = roc_curve(y_test, y_prob_l1)
auc_score_l1 = roc_auc_score(y_test, y_prob_l1)

plt.figure(figsize=(10, 6))
plt.plot(fpr_l1, tpr_l1, color='darkorange', label=f'ROC Curve (AUC = {auc_score_l1:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve with L1 Regularization')
plt.legend(loc='lower right')
plt.show()
```
